````python
from sklearn.linear_model import LinearRegression

# 'ì—¬ìë¹„ìœ¨'ì´ NaNì¸ 'êµ¬ë¶„2'ì˜ ê³ ìœ ê°’ì„ ì‹ë³„í•©ë‹ˆë‹¤. ì´ì „ ì§€ì‹œì™€ ë‹¬ë¦¬, ì—¬ê¸°ì„œëŠ” ì—¬ìë¹„ìœ¨ì´ NaNì¸ ê²½ìš°ì— ì§‘ì¤‘í•©ë‹ˆë‹¤.
null_women_ratio_groups = df_target[df_target['ì—¬ìë¹„ìœ¨'].isnull()]['êµ¬ë¶„2'].unique()

results = []

# 'êµ¬ë¶„2' ê°’ë“¤ì„ ê¸°ì¤€ìœ¼ë¡œ ë¶„ì„ì„ ì§„í–‰í•©ë‹ˆë‹¤. ì—¬ìë¹„ìœ¨ì´ NaNì¸ ê²½ìš°ê°€ ì•„ë‹ˆë¼, í•´ë‹¹ êµ¬ë¶„2ì— ëŒ€í•œ ì „ì²´ ë°ì´í„°ë¥¼ ë¶„ì„í•©ë‹ˆë‹¤.
for êµ¬ë¶„2 in null_women_ratio_groups:
    group = df_target[df_target['êµ¬ë¶„2'] == êµ¬ë¶„2]
    
    # ì—¬ìë¹„ìœ¨ì´ NaNì´ ì•„ë‹Œ ë°ì´í„°ë§Œ í•„í„°ë§í•˜ì—¬ ìœ íš¨í•œ ë¶„ì„ ë°ì´í„°ì…‹ì„ ìƒì„±í•©ë‹ˆë‹¤.
    valid_women_ratio_group = group.dropna(subset=['ì—¬ìë¹„ìœ¨'])

    if not valid_women_ratio_group.empty:
        # ê¸°ê°„ì„ ë…ë¦½ ë³€ìˆ˜ë¡œ, ì—¬ìë¹„ìœ¨ì„ ì¢…ì† ë³€ìˆ˜ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
        x = valid_women_ratio_group['ê¸°ê°„'].values.reshape(-1, 1)  # ê¸°ê°„ì„ 2D ë°°ì—´ë¡œ ë³€í™˜
        y = valid_women_ratio_group['ì—¬ìë¹„ìœ¨'].values  # ì—¬ìë¹„ìœ¨ì„ ì¢…ì† ë³€ìˆ˜ë¡œ ì‚¬ìš©

        # LinearRegression ëª¨ë¸ì„ ì´ˆê¸°í™”í•˜ê³  í•™ìŠµí•©ë‹ˆë‹¤.
        model = LinearRegression().fit(x, y)

        # íšŒê·€ê³„ìˆ˜ (ê¸°ìš¸ê¸°) ê³„ì‚°
        íšŒê·€ê³„ìˆ˜ = model.coef_[0]

        # MinMax, í‰ê· , í‘œì¤€í¸ì°¨ ê³„ì‚° (ëª¨ë‘ ì ˆëŒ€ê°’ìœ¼ë¡œ ì²˜ë¦¬)
        MinMax = round(abs(y.max() - y.min()), 6)
        í‰ê·  = round(abs(np.mean(y)), 6)
        í‘œì¤€í¸ì°¨ = round(abs(np.std(y, ddof=1)), 6)  # í‘œë³¸ í‘œì¤€í¸ì°¨ ê³„ì‚°

        # ê³„ì‚°ëœ ê°’ì„ ê²°ê³¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€í•©ë‹ˆë‹¤.
        results.append({
            'êµ¬': êµ¬ë¶„2,
            'íšŒê·€ê³„ìˆ˜': round(íšŒê·€ê³„ìˆ˜, 6),
            'MinMax':  MinMax,
            'í‰ê· ': í‰ê· ,
            'í‘œì¤€í¸ì°¨': í‘œì¤€í¸ì°¨
        })

# ê²°ê³¼ DataFrame ìƒì„± ë° 'êµ¬' ì»¬ëŸ¼ì„ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬
df_result = pd.DataFrame(results).sort_values(by='êµ¬').reset_index(drop=True)
````
