#### 모듈로 수정할것

```python
# 출연진 이름 빼기

#성별

m_data=[] #남성
w_data=[] #여성

#나이

child=[] #어린이
teen=[] #청소년
you=[] #청년
mid=[] #중년
o=[] #노인

#직업

o1 = [] #전문기술
o2 = [] #연예인
o3 = [] #기타
o4 = [] #서비스
o5 = [] #운동선수
o6 = [] #사회복지
o7 = [] #예술가
o8 = [] #산업
o9 = [] #금융




for i in range(229):
    male_count = 0
    female_count = 0
    children = 0
    Teenager = 0
    youth = 0
    middle =0
    old =0 

    job1=0 # 모듈로 나중에 수정
    job2=0
    job3=0
    job4=0
    job5=0
    job6=0
    job7=0
    job8=0
    job9=0


    actors = new_data['출연자'][i].split(' ')
    for name in actors:
        filtered_df = name_data[name_data['이름'] == name]
        if not filtered_df.empty:
            gender = filtered_df['성별'].iloc[0]  
            print(f'gender :{gender}')
            if gender == 'M':
                male_count += 1
            elif gender == 'W':
                female_count += 1
        if not filtered_df.empty:
            age = filtered_df['나이'].iloc[0]
            if age == '어린이':
                children +=1
            elif age == '청소년':
                Teenager +=1
            elif age == '청년':
                youth +=1
            elif age == '중년':
                middle +=1
            elif age == '노인':
                old +=1
        if not filtered_df.empty:
            job = filtered_df['직업'].iloc[0]
            #전문기술(1)', '연예인(2)', '기타(3)', '서비스(4)', '운동선수(5)', '사회복지(6)', '예술가(7)', '산업(8)', '금융(9)'
            if job == '전문기술':
                job1 +=1
            elif job == '연예인':
                job2 +=1
            elif job == '기타':
                job3 +=1
            elif job == '서비스':
                job4 +=1
            elif job == '운동선수':
                job5 +=1
            elif job == '사회복지':
                job6 +=1
            elif job == '예술가':
                job7 +=1
            elif job == '산업':
                job8 +=1
            elif job == '금융':
                job9 +=1



    m_data.append(male_count)
    w_data.append(female_count)
    child.append(children)
    teen.append(Teenager)
    you.append(youth)
    mid.append(middle)
    o.append(old)

    o1.append(job1)
    o2.append(job2)
    o3.append(job3)
    o4.append(job4)
    o5.append(job5)
    o6.append(job6)
    o7.append(job7)
    o8.append(job8)
    o9.append(job9)



    print(f"{i}회차")

    print(male_count)
    print(female_count)
    print("-----------------")

print(f"Male count per episode: {m_data}")
print(f"Female count per episode: {w_data}")

        
      
```
