```python
import pandas as pd
import re

def extract_names(names_str):
    # 대괄호([])와 그 안의 내용을 제거
    names_clean = re.sub(r'\[\d+\]', '', names_str)
    # ', '와 ' & '로 문자열을 분리하여 리스트로 만듦
    names_list = re.split(r',\s*|\s*&\s*', names_clean)
    return names_list

# 예시 DataFrame 생성 (실제 데이터에 맞게 조정 필요)
# df_name = pd.DataFrame({'회차': ['1', '2'], '출연자_정보': ['name1[1], name2 & name3[2]', 'name4, name5 & name6']})
# df = pd.DataFrame({'제목': ['EP.1 something', 'EP.2 something else'], '출연자_정보': ['', '']})

# '출연자_정보' 컬럼에 대해 함수 적용
for i in range(len(df_name)):
    names_str_example = df_name['출연자_정보'][i]
    extracted_names = extract_names(names_str_example)

    num = df_name['회차'][i]
    mask = df['제목'].str.contains(f'EP.{num}', case=False, na=False)
    
    df.loc[mask, '출연자_정보'] = ', '.join(extracted_names)  # 리스트를 문자열로 변환하여 저장
    print(extracted_names)  # 결과 출력
```
