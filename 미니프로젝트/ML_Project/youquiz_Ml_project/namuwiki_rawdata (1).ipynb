```python
import pandas as pd
import re

# 웹 크롤링에 사용할 라이브러리
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chrome_driver_path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(executable_path=chrome_driver_path)
url = "https://namu.wiki/w/%EC%9C%A0%20%ED%80%B4%EC%A6%88%20%EC%98%A8%20%EB%8D%94%20%EB%B8%94%EB%9F%AD/%EB%B6%80%EC%A0%9C%20%EB%B0%8F%20%EC%8B%9C%EC%B2%AD%EB%A5%A0"
driver.get(url)

# DataFrame을 만들기 위한 빈 리스트 생성
data = {'시즌': [], '에피소드': [], '날짜': [], '주제': [], '출연자': [], '직업': [], '시청률': [], '인지도': []}

# 테이블의 행만큼 반복
for row_index in range(2, 14):
    # WebDriverWait를 사용하여 페이지 이동 대기
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[1]/div/strong'))
        )
    except:
        break

    # XPath를 사용하여 각 항목의 데이터 추출
    episode = driver.find_element(By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[1]/div/strong')
    date = driver.find_element(By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[2]')
    topic = driver.find_element(By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[3]')
    cast = driver.find_element(By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[4]')
    cast_elements = driver.find_elements(By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[4]/div/a')
    rating = driver.find_element(By.XPATH, f'//*[@id="app"]/main/div[2]/div/article/div[4]/div/div[2]/div/div/div[6]/div[1]/table/tbody/tr[{row_index}]/td[6]')

    # 특정 열의 모든 값에서 숫자, 기호 및 쉼표를 제외한 문자만 추출
    cast_text_1 = ''.join(char for char in cast.text if char.isalpha() or char.isspace())

    # '출연자' 열에서 기호 및 쉼표 제외하고 숫자만 개별로 출력
    cast_text_2 = ', '.join(re.findall(r'\d+', cast.text))

    # '시청률' 열에서 '%' 제거 및 소수점 유지
    rating_text = re.sub(r'%', '', rating.text)

   # '인지도' 열에 링크가 있는지 여부에 따라 값 추가
    if any(cast_elements):
        data['인지도'].append('y')
    else:
        data['인지도'].append('n')

    # 추출한 데이터를 리스트에 추가
    data['시즌'].append(1)  # 시즌 값 추가
    data['에피소드'].append(episode.text)
    data['날짜'].append(date.text)
    data['주제'].append(topic.text)
    data['출연자'].append(cast_text_1)
    data['직업'].append(cast_text_2)
    data['시청률'].append(rating_text)

# DataFrame 생성
df = pd.DataFrame(data)

# 브라우저 닫기
driver.quit()

# '날짜' 열을 datetime 형식으로 변경
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y년 %m월 %d일')

# '주제' 컬럼에서 괄호와 그 안의 숫자를 삭제
df['주제'] = df['주제'].replace(to_replace=r'\s*\[.*?\]\s*', value='', regex=True)

# DataFrame 출력
df_cast_elements_1 = df
df_cast_elements_1
```
